```python
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Specify the path to your file in Google Drive
file_path = '/content/drive/My Drive/BLV/Forecast/RAW_DATA/weather_data.csv'

# Load the data
data = pd.read_csv(file_path, sep=',')

# Display the data
print(data.head())
```

```python
data.describe(include = 'all')
```

```python
data.fillna(method='ffill', inplace=True)  # điền bằng giá trị trước đó
```

```python
data.describe(include = 'all')
```

```python
# Đảm bảo cột Time là datetime
data['Time'] = pd.to_datetime(data['Time'])

# Set Time làm chỉ số (index)
data = data.set_index('Time')

# Resample về tần suất giờ (hourly), dùng trung bình cho các giá trị numeric
data_hourly = data.resample('H').mean()
```

```python
data_hourly.describe(include = 'all')
```

```python
# Trả lại cột Time
data_hourly = data_hourly.reset_index()

# Lưu file CSV
data_hourly.to_csv('resampled_hourly.csv', index=False)
```
